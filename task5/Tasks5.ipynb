```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Category': ['l', 'l', 'd', 'd', 'd', 'h', 'h', 'h', 'h'],
    'Values': [100, 200, 300, 400, 500, 600, 700, 800, 900]
}
df = pd.DataFrame(data)

def calculate_range(series):
    return series.max() - series.min()

result = df.groupby('Category')['Values'].agg(calculate_range)

fig, ax = plt.subplots(figsize=(8, 5))
result.plot(kind='bar', color='skyblue', ax=ax)
ax.set_title("Custom Aggregation: Range by Category", fontsize=14)
ax.set_xlabel("Category", fontsize=12)
ax.set_ylabel("Range (Max - Min)", fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```
